```python
def verificar_longitud(palabra, numero):
    if len (palabra) == numero:
        return True
    else:
        return False

verificar_longitud('hola' , 6)
```

```python
def revisar_letra(palabra, letra):
    if letra in palabra:
        return f"[{letra}]" 
    else:
        return letra          
revisar_letra('maria' , 'i')


    
```
